# Netflix Data Analysis: 1990s Movies

This notebook explores Netflix's movie data to analyze trends in the 1990s. Specifically, it focuses on identifying and analyzing **short action movies** (less than 90 minutes) from that decade.

## Objective
- Explore the dataset and extract insights about movies from the 1990s.
- Identify the number of short action movies from that decade.

### Dataset: `netflix_data.csv`
Below is a brief description of the dataset used:

| Column         | Description                       |
|----------------|-----------------------------------|
| `show_id`      | Unique identifier for the show   |
| `type`         | Type of content (Movie/TV Show)  |
| `title`        | Title of the show                |
| `director`     | Director of the show             |
| `cast`         | Cast members                     |
| `country`      | Country of origin                |
| `date_added`   | Date added to Netflix            |
| `release_year` | Year of release                  |
| `duration`     | Duration in minutes              |
| `description`  | Brief description of the show    |
| `genre`        | Genre of the show                |


```python
# Step 1: Importing required libraries
import pandas as pd
import matplotlib.pyplot as plt

# Step 2: Load the Netflix dataset
netflix_df = pd.read_csv("netflix_data.csv")

# Display the first few rows
netflix_df.head()
```

```python
# Step 3: Filter movies released between 1990 and 1999
movies_1990s = netflix_df[(netflix_df['release_year'] >= 1990) & (netflix_df['release_year'] <= 1999)]
movies_1990s.head()
```

```python
# Step 4: Visualize the distribution of movies by release year
movies_1990s['release_year'].hist(bins=10, color='skyblue', edgecolor='black')
plt.title("Distribution of Movies by Release Year (1990-1999)")
plt.xlabel("Release Year")
plt.ylabel("Count")
plt.show()
```

```python
# Step 5: Filter short action movies (< 90 minutes)
short_action_movies = movies_1990s[
    (movies_1990s['duration'] < 90) & (movies_1990s['genre'] == 'Action')
]

# Display the filtered dataset
short_action_movies
```

```python
# Step 6: Count the number of short action movies
short_movie_count = len(short_action_movies)
print(f"Number of short action movies from the 1990s: {short_movie_count}")
```
